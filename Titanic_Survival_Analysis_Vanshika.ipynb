```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(pd.read_csv('/content/train (1).xls'))
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
drop_col = df.isnull().sum()[df.isnull().sum()>(35/100 * df.shape[0])]
drop_col
```

```python
drop_col.index
```

```python
df.fillna(df.mean(), inplace=True)
df.isnull().sum()
```

```python
df['Embarked'].describe()
```

```python
df['Embarked'].fillna("S", inplace=True)
df.isnull().sum()
```

```python
df.corr()
```

```python
df['FamilyMan'] = df['SibSp'] + df['Parch']
df.drop(['SibSp', 'Parch'], axis=1, inplace=True)
df.corr()
```

```python
df['SinglePerson'] = [0 if df['FamilyMan'][i]>0 else 1 for i in df.index]
df.head()
```

```python
df.groupby(['SinglePerson'])['Survived'].mean()
```

```python
df[['SinglePerson', 'Fare']].corr()
```

```python
df['Sex'] = [0 if df['Sex'][i]=='male' else 1 for i in df.index] 
#Note: 0 is for Male and 1 is for FeMale
df.groupby(['Sex'])['Survived'].mean()
```

```python
df.groupby(['Embarked'])['Survived'].mean()
```
